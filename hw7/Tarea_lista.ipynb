```python
#Definición de las condiciones iniciales y de contorno

u_pasado = 3.0 * sin(pi*x)
u_pasado[where((x==0)& (x==2))]= 0 
u_todos[:,1] = u_pasado[:]

u_pasado_pasado = u_pasado.copy()
u_pasado_pasado[where((x==0)& (x==2))]= 0
u_todos[:,0] = u_pasado_pasado[:]

fig = plt.figure(figsize(15,10)) #Definiendo el tamaño de todas las figuras
plot(x,u_pasado) #Imprimir la condición inicial
title('Condicion Inicial')
xlabel('Posicion X')
ylabel('velocidad U')
```

```python
#Cuando el tiempo transcurrido es T = 0.15

for j in range(2,puntos_t):
    for i in range(1,num_puntos-1):
        u_presente[i] = u_pasado_pasado[i] - beta*(((u_pasado[i+1])**2 - (u_pasado[i-1])**2)/2.0)
    u_pasado_pasado[:] = u_pasado[:]
    u_pasado[:] = u_presente[:]
    u_todos[:,j] = u_presente[:]
    plot(u_todos[:,j])
    #plt.show()
```
